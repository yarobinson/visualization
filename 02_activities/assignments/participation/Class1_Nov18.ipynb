```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import seaborn as sns
```

```python
np.random.seed(42)
x = np.arange(50)
y = np.random.randint(0,100,50) #first value is our minimum, second value is our max, third value is the number of data points

fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.hist(y)
```

```python
font1 = {'family': 'sans-serif', 'color' : 'blue', 'size':20}
font2 = {'family': 'monospace', 'color' : 'green', 'size':14}

fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict= font1)
ax.set_ylabel('Total Growth', fontdict= font2)
ax.set_xlabel('Years since start', fontdict= font2)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
           marker = '*',
           color = 'indigo')
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
           marker = '*',
           color = 'indigo',
           linestyle = '--',
           linewidth = 2)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
           marker = '*',
           color = 'indigo',
           linestyle = '--',
           linewidth = 2,
           markeredgecolor = '#FA9359',
           markerfacecolor = '#000000')
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x,y,
           marker = '*',
           color = 'indigo',
           linestyle = '--',
           linewidth = 2,
           markeredgecolor = '#FA9359',
           markerfacecolor = '#000000')

ax.grid(axis = 'y')
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x,y,
           marker = '^',
           color = 'seagreen',
           linewidth = 1,
           markeredgecolor = 'black',
           markerfacecolor = '#000000')

ax.grid(axis = 'y',
        color = 'red')
```
